```python
import pymongo
import pandas as pd

# Cadena de conexión a MongoDB
mongo_connection_string = "mongodb://admin:admin@localhost:27017/"

# Conexión a MongoDB
client = pymongo.MongoClient(mongo_connection_string)

# Seleccionar la base de datos 'valenbici'
db = client['valenbici']

# Seleccionar la colección 'sample_table'
collection = db['sample_table']

# Obtener todos los documentos de la colección
documents = list(collection.find())

# Crear un DataFrame a partir de la lista de documentos
df = pd.DataFrame(documents)

# Cierra la conexión
client.close()

df['updated_at'] = pd.to_datetime(df['updated_at'], format='%d/%m/%Y %H:%M:%S')


df
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Suponiendo que 'df' es tu DataFrame
df['updated_at'] = pd.to_datetime(df['updated_at'], format='%d/%m/%Y %H:%M:%S')

# Agrupar por tiempo y sumar la disponibilidad
total_disponibilidad_por_tiempo = df.groupby('updated_at')['available'].sum().reset_index()

# Crear un gráfico de líneas con la suma total de disponibilidad por tiempo
plt.figure(figsize=(14, 8))
plt.plot(total_disponibilidad_por_tiempo['updated_at'], total_disponibilidad_por_tiempo['available'], marker='o')

plt.title('Suma Total de Disponibilidad a lo largo del Tiempo')
plt.xlabel('Fecha')
plt.ylabel('Total de Disponibilidad')
plt.show()
```

```python
import plotly.express as px

# Suponiendo que 'updated_at' es la columna de fechas
df['updated_at'] = pd.to_datetime(df['updated_at'])
df = df.sort_values(by='updated_at')  # Ordenar por fecha

fig = px.bar(df, x='updated_at', y='available', labels={'available': 'Disponibilidad'}, title='Evolución de Disponibilidad a lo largo del Tiempo')
fig.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Suponiendo que 'df' es tu DataFrame
df['updated_at'] = pd.to_datetime(df['updated_at'], format='%d/%m/%Y %H:%M:%S')

# Crear un gráfico de líneas para cada dirección
plt.figure(figsize=(14, 8))
for address, group in df.groupby('address'):
    group = group.sort_values(by='updated_at')  # Ordenar por fecha
    plt.plot(group['updated_at'], group['available'], label=address, marker='o')

plt.title('Evolución de Disponibilidad por Dirección a lo largo del Tiempo')
plt.xlabel('Fecha')
plt.ylabel('Disponibilidad')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))  # Coloca la leyenda a la derecha del gráfico
plt.tight_layout()  # Ajusta automáticamente la disposición para evitar superposiciones
plt.show()

```
